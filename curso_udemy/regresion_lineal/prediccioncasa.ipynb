```python
from sklearn.model_selection import train_test_split
X = df.loc[:,["OverallQual","GrLivArea","GarageCars"
                        ,"TotalBsmtSF","YearBuilt","YearRemodAdd"]]
y = df.loc[:,['SalePrice']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=33)

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
predicciones = lm.predict(X_test)
Dtpredicciones = pd.DataFrame(predicciones)

#Resetear los index para unir los 2 dataFrame  y_test y dtPredicciones
Dtpredicciones.reset_index(drop = True, inplace = True) 
y_test.reset_index(drop = True, inplace = True)
df_unido = y_test.join(Dtpredicciones)

df_unido.columns['SalePrice', 'Predic']
```

# Metricas

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test,predicciones))
print('MAE:', metrics.mean_squared_error(y_test,predicciones))
print('MAE:', np.sqrt(metrics.mean_squared_error(y_test,predicciones)))
```

# Generar el modelo predictivo de regresion lineal

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

test = pd.read_csv("./data/test.csv")

X = test.loc[:,["OverallQual","GrLivArea","GarageCars"
                        ,"TotalBsmtSF","YearBuilt","YearRemodAdd"]]
```

Limpiando y verificando los NA del DataSet

```python
X.isna().sum() # Todos los que tienen un 1 tiene un NA
```

```python
X['GarageCars'].fillna(0, inplace = True) # Remplazar los NA en 0
X['TotalBsmtSF'].fillna( X['TotalBsmtSF'].mean() , inplace = True) # Remplazar los NA en media
```

Volvemos a comprobar que ya no haya NA

```python
X.isna().sum()
```

```python
predicciones = lm.predict(X)
```

```python
Dtpredicciones = pd.DataFrame(predicciones) #transforma los datos predicciones a un DataFrame
id = test.loc[:,['Id']] #Extraer el ID del datafram principal

Dtpredicciones.reset_index(drop = True, inplace = True)
id.reset_index(drop = True, inplace = True)

# Uniendo los datos
df_entrega = id.join(Dtpredicciones)
df_entrega.columns = ['id', 'SalePrice']

df_entrega.head(5)
```

```python
df_entrega.to_csv('entrega.csv', index = False)
```

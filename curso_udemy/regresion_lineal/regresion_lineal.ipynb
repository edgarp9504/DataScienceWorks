```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
iris = sb.load_dataset("iris")
iris.head()
```

```python
iris.describe()
```

```python
sb.pairplot(iris)
```

```python
from sklearn.model_selection import train_test_split
X = iris[['petal_length']]
y = iris[['petal_width']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=33)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
predicciones = lm.predict(X_test)
predicciones
```

```python
plt.scatter(y_test, predicciones)
plt.plot([0,2.5], [0,2.5])
plt.show()
```

```python
plt.scatter(X_test, predicciones)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test,predicciones))
print('MAE:', metrics.mean_squared_error(y_test,predicciones))
print('MAE:', np.sqrt(metrics.mean_squared_error(y_test,predicciones)))
```
